```python
df1 = df[df['total_shootings'] <200]
df1.boxplot(column=['total_shootings'])

```

```python
table2 = pd.pivot_table(df, columns = ['type'], values= 'total_shootings', aggfunc = 'sum', fill_value = 0)
table2
```

```python
import numpy as np

plt.figure(figsize=(10,5))

plt.bar(table2.columns[0:], table2.iloc[0] ) #data for the plot

plt.title("Police Agency shooting's")
plt.xlabel('Police Agencies')
plt.ylabel('# of Shootings')
plt.legend(["Fatal Shootings"], loc = "upper right")

plt.yticks(np.arange(0, 6000, 1000))  # the numbers on the y axis

plt.show()
```

The graph shows the number of fatal shootings by the different Police Agencies. Fatal shootings by the local police is drastically higher than the other police agencies with the sheriff agency second. The local police have a total of 5690 fatal shootings, almost doubling the size of the sheriff and tripling the other agencies. The local police(5690) and sheriff(2668) have the highest fatal shooting numbers among the different agencies, leaving a huge gap between them and the rest of the agencies. Local police have more interaction with civilians when responding to crimes; however, their high degree of killing indicate a lack of training within the areas of use of force and rules of engagement. The "other" agency is the ideal agency with an impressive low number of fatal shootings. The "other" agency law enforcement training is proven to be effective, beneficial, and life preserving. Overall, the local police fatal shooting numbers are dangerously high and proves a lack of effective training among the different agencies.

```python
table = pd.pivot_table(df, index = ['state'], values= 'total_shootings', aggfunc = 'sum' ,fill_value = 0)
table
```

```python
tbl1 = table.sort_values(by = 'total_shootings', ascending = True)
tbl1
```

```python
plt.figure(figsize=(10,10))

plt.barh(tbl1.index, tbl1.iloc[:,0]  ) 



plt.title("# of Police Shooting's in Different State")
plt.xlabel('# of Shootings')
plt.ylabel("State's")
plt.legend(['Fatal Shootings'], loc = "lower right")

plt.xticks(np.arange(0, 1400, 100))  # the numbers on the y axis

plt.show()
```

The graph examines the number of fatal shootings by state. California has an extremely high number of fatal shootings. California have a total of 1282 fatal shootings. California compared to the US differ drastically, being the US represents a select few of federal agencies. Federal agencies rank among the few agencies with low number of fatal shootings. Rhode Island along with the US has 10 killings. Rhode Island is the smallest state by area and California is one of the most populated states in the US. States within the same range of area and population have a similar number of fatal shootings. Indiana and Illinois are similar in population and area with 185 fatal shootings. Overall, the graph show that the most populated states have the most fatal shooting numbers and the smallest states have the least amount.

```python
new_table = pd.pivot_table(df, index = ['name'], values= 'total_shootings', aggfunc = 'sum' ,fill_value = 0)
new_tbl = new_table.sort_values(by = 'total_shootings', ascending = False)
new_tbl
```

```python
pie_df = new_tbl.head(10)
pie_df
```

```python
plt.pie(pie_df.iloc[:, 0], explode = [0.2, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05], labels = pie_df.index, shadow = True, autopct = '%.2f%%')
plt.title('Top 10 Police Departments with the highest Fatal Shooting Percentage')
plt.show()
```

The pie chart exemplifies the percentage of the top 10 police departments with the highest fatal shootings. Los Angelas Police Department has the highest percentage with 15.56%. The top 10 police departments percentage's drop steadily rather than rapidly, displaying the number of fatal shootings acquired by all of states decreases in small amounts. California's Highway Patrol has the lowest percentage with 6.51%. Highway Patrol occupies the highway and encounter less interaction than Los Angelas Police Department. Majority of the top 10 police departments occupies the largest states. The pie chart shows a steady declining percentage of fatal shootings through the top 10 departments.
